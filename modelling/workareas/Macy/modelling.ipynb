```python
ROOT = path.abspath(curdir)

datapath = path.join(ROOT, 'data', 'calculations.csv')

calc = pd.read_csv(datapath)
calc  = calc.iloc[:,1:]
calc.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = ["TotalViolentCrimes", "TotalNonviolentCrimes", "OtherCrimes"]
target = "RecidivismRate"

X = calc[features]
y = calc[target]

# Normalize features
#scaler = StandardScaler()
#X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8)
```

```python
from sklearn import linear_model as lm
model = lm.LinearRegression()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
treeModel = RandomForestRegressor()
treeModel.fit(X_train, y_train)
treeModel.score(X_test,y_test)
```
